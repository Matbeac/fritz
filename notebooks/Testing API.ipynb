## Spoonacular

```python
import requests

url = "https://spoonacular-recipe-food-nutrition-v1.p.rapidapi.com/recipes/search"

querystring = {"query":"cheesecake","number":"100","type":"main course"}

headers = {
    'x-rapidapi-host': "spoonacular-recipe-food-nutrition-v1.p.rapidapi.com",
    'x-rapidapi-key': "312fd0219amsh6881253b288c17dp1825c8jsn47f925e37c6f"
    }
```

```python
response = requests.request("GET", url, headers=headers, params=querystring)
```

```python
response.json()
```

```python
# Recipe name

recipe = response.json()['results'][0]['title']
recipe
```

```python
# ID recipe, neccesary for next request

id_recipe = response.json()['results'][0]['id']
```

```python
# Number of servings 

servings = response.json()['results'][0]['servings']
servings
```

```python
# Getting ingredients

import requests

url = f"https://spoonacular-recipe-food-nutrition-v1.p.rapidapi.com/recipes/{id_recipe}/ingredientWidget.json"

headers = {
    'x-rapidapi-host': "spoonacular-recipe-food-nutrition-v1.p.rapidapi.com",
    'x-rapidapi-key': "312fd0219amsh6881253b288c17dp1825c8jsn47f925e37c6f"
    }
```

```python
response2 = requests.request("GET", url, headers=headers)
```

```python
# All the ingredients 

ingredients = response2.json()['ingredients']
ingredients
```

```python
for i in ingredients:
    print(i)
```

```python
# Ingredient name

ingredients[0]['name']
```

```python
# Quantity per portion (1 person)

ingredients[0]['amount']['metric']['value'] / servings
```

```python
# Metric

ingredients[0]['amount']['metric']['unit']
```
